```python
df["timestamp"] = pd.to_datetime(
    df["timestamp"],
    format="mixed"
)
```

```python
df["timestamp"] = pd.to_datetime(
    df["timestamp"],
    format="mixed",
    errors="coerce"
)
```

```python
df["timestamp"].isna().sum()
```

```python
df["hour"] = df["timestamp"].dt.hour
df["day_of_week"] = df["timestamp"].dt.dayofweek
df["month"] = df["timestamp"].dt.month
df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)
```
